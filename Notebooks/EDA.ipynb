# 5. Check for any patterns or trends

This output shows the average rating given by each user. Each line corresponds to one user, identified by the reviewerID column. The value to the right of each user ID is the average of all ratings that this user has given. A value of 5.0 means that the user has given a 5-star rating on average, which is the highest possible rating. Conversely, a user with an average rating of 1.0 has given the lowest possible rating on average.

```python
average_rating_per_user = reviews_df.groupby('reviewerID')['overall'].mean().sort_values(ascending=False)
print(average_rating_per_user)
```

# 8. Analyzing review texts

```python
# Create a new column for the review length
reviews_df['review_length'] = reviews_df['reviewText'].apply(lambda x: len(str(x)))

# Plotting length of reviews
plt.figure(figsize=(10,6))
reviews_df['review_length'].plot(bins=50, kind='hist', color='blue', edgecolor='black')
plt.xlabel('Review Length', fontsize=12)
plt.title('Distribution of Review Length', fontsize=12)
plt.show()

# Correlation between review length and rating
print("Correlation between review length and rating: ", reviews_df['overall'].corr(reviews_df['review_length']))
```

# 10. Analyzing distribution of ratings

How have the number of reviews changed over time? Have they been increasing or decreasing? What are the most active months for reviews? To answer these questions, we can convert the reviewTime column to datetime and analyze the number of reviews per year/month.

```python
# Convert reviewTime to datetime
reviews_df['reviewTime'] = pd.to_datetime(reviews_df['reviewTime'])

# Extract year and month
reviews_df['year'] = reviews_df['reviewTime'].dt.year
reviews_df['month'] = reviews_df['reviewTime'].dt.month

# Count reviews per year
reviews_per_year = reviews_df['year'].value_counts().sort_index()

# Count reviews per month
reviews_per_month = reviews_df['month'].value_counts().sort_index()

# Plotting
fig, ax = plt.subplots(1, 2, figsize=(15,5))
sns.lineplot(x=reviews_per_year.index, y=reviews_per_year.values, ax=ax[0])
sns.barplot(x=reviews_per_month.index, y=reviews_per_month.values, ax=ax[1], palette='viridis')
ax[0].set_xlabel('Year', fontsize=12)
ax[0].set_ylabel('Count', fontsize=12)
ax[0].set_title('Reviews per Year', fontsize=12)
ax[1].set_xlabel('Month', fontsize=12)
ax[1].set_ylabel('Count', fontsize=12)
ax[1].set_title('Reviews per Month', fontsize=12)
plt.show()
```

# 11. Analyzing metadata

```python
# Count reviews per brand
brand_counts = metadata_df['brand'].value_counts()

# Get top 10 brands by reviews
top_brands = brand_counts[:10]

# Plotting
plt.figure(figsize=(10,6))
sns.barplot(x=top_brands.index, y=top_brands.values, palette='viridis')
plt.xticks(rotation=90)
plt.xlabel('Brands', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Top 10 Brands by Reviews', fontsize=12)
plt.show()
```

# 12. Plot the distribution of ratings.

```python
# Plotting
plt.figure(figsize=(10,6))
sns.countplot(x='overall', data=reviews_df, palette='viridis')
plt.xlabel('Ratings', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Distribution of Ratings', fontsize=12)
plt.show()
```

# 13. Look at the number of reviews over time, grouped by ratings.

This will give us an idea if there's a particular trend in ratings over time.

```python
# Convert reviewTime to datetime
reviews_df['reviewTime'] = pd.to_datetime(reviews_df['reviewTime'])

# Group by year and ratings
reviews_over_time = reviews_df.groupby([reviews_df.reviewTime.dt.year, 'overall']).size().unstack(fill_value=0)

# Plotting
reviews_over_time.plot(kind='bar', stacked=True, colormap='viridis', figsize=(10,6))
plt.xlabel('Year', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Distribution of Ratings Over Time', fontsize=12)
plt.show()
```
